```python
import PyOctaveBand
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample rate and duration
fs = 48000
duration = 5  # In seconds
frq_limits = [12, 20000]
```

```python
# Time array
x = np.arange(np.round(fs * duration)) / fs
print(x)
```

```python
def signal(x):
    # Signal with 6 frequencies
    f1, f2, f3, f4, f5, f6 = 20, 100, 500, 2000, 4000, 15000
    # Multi Sine wave signal
    return 100 \
        * (np.sin(2 * np.pi * f1 * x)
           + np.sin(2 * np.pi * f2 * x)
           + np.sin(2 * np.pi * f3 * x)
           + np.sin(2 * np.pi * f4 * x)
           + np.sin(2 * np.pi * f5 * x)
           + np.sin(2 * np.pi * f6 * x))
```

```python
y = signal(x)

# Filter (only octave spectra)
spl, freq = PyOctaveBand.octavefilter(y, fs=fs, fraction=3, order=6, limits=[12, 20000])

plt.figure(figsize=(20,8))
plt.plot(freq, spl)
plt.xscale('log')
plt.grid()
plt.show()
```
